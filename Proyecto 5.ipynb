SE CALCULA EL INGRESO MENSUAL POR USUARIO, OBTENIENDO EL MONTO POR MINUTO DE LLAMADA, MENSAJES Y GB EXCEDIDOS Y SUMANDO A LA TARIFA BASE.

```python

df_integrado["minutos_excedidos"]=df_integrado["calls_duration"]-df_integrado["minutes_included"]# Calcula el ingreso mensual para cada usuario
df_integrado["mensajes_excedidos"]=df_integrado["messages_send"]-df_integrado["messages_included"]
df_integrado["internet_excedidos"]=df_integrado["mb_used"]-df_integrado["mb_per_month_included"]
```

```python

df_integrado["cargos_min"] = np.where(
    df_integrado["minutos_excedidos"] > 0,
    df_integrado["minutos_excedidos"] * df_integrado["usd_per_minute"],
   
 0
)
df_integrado["cargos_mensaje"] = np.where(
    df_integrado["mensajes_excedidos"] > 0,
    df_integrado["mensajes_excedidos"] * df_integrado["usd_per_message"],
   
 0
)
# Convertir MB excedidos a GB (redondeado hacia arriba)
df_integrado["gb_excedidos"] = np.where(
    df_integrado["internet_excedidos"] > 0,
    np.ceil(df_integrado["internet_excedidos"] / 1024),  # Usar 1024
    0
)
df_integrado["cargos_internet"] = np.where(
    df_integrado["gb_excedidos"] > 0,
    df_integrado["gb_excedidos"] * df_integrado["usd_per_gb"],0)
```

```python

df_integrado["ingreso_total"]= df_integrado["cargos_min"]+df_integrado["cargos_mensaje"]+df_integrado["cargos_internet"]+df_integrado["usd_monthly_pay"]
```

## Prueba las hipótesis estadísticas


H0: No hay diferencia significativa entre los ingresos promedio procedentes de los planes surf y ultimate.
Ha: Los ingresos promedio procedentes de los planes surf y ultimate con diferentes.


```python

alpha=0.05
surf= df_integrado[df_integrado["plan_name"]=="surf"]["ingreso_total"]
ultimate= df_integrado[df_integrado["plan_name"]=="ultimate"]["ingreso_total"]
result=st.ttest_ind(surf,ultimate,equal_var=False)
print('valor p:', result.pvalue) 
if (result.pvalue<alpha):# su código: compara los valores p obtenidos con el nivel de significación estadística):
    print("Rechazamos la hipótesis nula")
else:
    print("No podemos rechazar la hipótesis nula")# Prueba las hipótesis


```

Ha: SI HAY DIFERENCIA SIGNIFICATIVA ENTRE LOS INGRESOS PROMEDIO DE USUARIOS DE NUEVA YORK-NUEVA JERSEY Y USUARIOS DE OTRAS REGIONES


```python
ny_nj=df_integrado[df_integrado["city"].isin(["New York-Newark-Jersey City, NY-NJ-PA MSA"])]["ingreso_total"]
```

```python
otras_regiones=df_integrado[~df_integrado["city"].isin(["New York-Newark-Jersey City, NY-NJ-PA MSA"])]["ingreso_total"]
```

```python
alpha=0.05
result=st.ttest_ind(ny_nj,otras_regiones,equal_var=False)
print('valor p:', result.pvalue) 
if (result.pvalue<alpha):# su código: compara los valores p obtenidos con el nivel de significación estadística):
    print("Rechazamos la hipótesis nula")
else:
    print("No podemos rechazar la hipótesis nula")
```

Es importante mencionar que el t-test asume que las dos muestras tienen varianzas iguales Para verificar esa igualdad de varianzas podemos usar el **test de Levene**. Este nos dice si la variabilidad entre los grupos es *estadísticamente diferente*. En futuros notebooks podrías utilizar este test para decidir sobre la igualdad de varianzas. En corto, el flujo sería así

1. Usamos `levene()` para verificar si las varianzas son iguales.
2. Dependiendo del resultado, usamos `ttest_ind()` con el argumento `equal_var`:
   - `equal_var=True` si Levene dice que las varianzas son iguales (p > 0.05).
   - `equal_var=False` si Levene dice que las varianzas son diferentes (p < 0.05).
